```python
import math 
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
t = np.arange(0,2*np.pi,0.01)
n = len(t)
y = np.zeros(n)
for i in range(n):
   if t[i] <= np.pi:
       y[i] = 1
   else:
       y[i] = -1
plt.plot(t,y)
n = 360
theta = np.linspace(0,2*np.pi,n+1)
attempts = np.array([5,10,100])
for i in range(3):
   l = attempts[i]
   z = np.zeros(n+1)
   for k in range(1,l,2):
      bk = 4 /np.pi/k
      z += bk*np.imag(np.exp(1.j*k*theta))
   plt.plot(theta,z,label = "N = " + str(l))
plt.plot(t,y)
plt.legend(loc='upper right',fontsize=10)
```
